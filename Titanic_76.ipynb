```python
from keras.utils import np_utils
import numpy as np
x_train = np.array(X_train)
y_train = np_utils.to_categorical(Y_train,  num_classes=2)
x_test = np.array(X_test)
```

```python
#validation
np.random.seed(1)
x_validation = list()
y_validation = list()
for i in range(90):
    ind = np.random.randint(0,x_train.shape[0])
    x_validation.append(x_train[ind])
    y_validation.append(y_train[ind])
    x_train = np.delete(x_train,ind,0)
    y_train = np.delete(y_train,ind,0)
x_validation = np.array(x_validation)
y_validation = np.array(y_validation)
```

```python
from keras.layers import Input, Dense,BatchNormalization,Dropout 
from keras.models import Model

inputs = Input(shape=(8,))

# a layer instance is callable on a tensor, and returns a tensor
output_1 = Dense(128, activation='selu' , kernel_initializer = 'he_uniform'  )(inputs)
output_1 = BatchNormalization()(output_1)
#output_1 = Dropout(rate = 0.2 , )(output_1)
output_2 = Dense(64, activation='selu' , kernel_initializer = 'he_uniform' , )(output_1)

output_2 = BatchNormalization()(output_2)


predictions = Dense(2, activation='softmax' , kernel_initializer ='he_uniform' )(output_2)

model = Model(inputs=inputs, outputs=predictions)

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
model.fit(x_train, y_train, verbose =2 ,validation_split = 0.1 ,epochs=200, batch_size=16,shuffle=True)

loss, accuracy = model.evaluate(x_train, y_train)
print(loss , accuracy)
```

```python
import csv

y_test = model.predict(x_test)
predict = list()
for j in y_test:
    if j[0]>j[1]:
        predict.append(0)
    else: 
        predict.append(1)
        
with open('/root/yaun/data/kaggle/DNN_3layers_BN_ep200_validation_output.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['PassengerId','Survived'])
    for i in range(len(predict)):
        writer.writerow([str(892+i),str(predict[i])]) #891 ~ 1308

# kaggel Score = 0.73205    
```
